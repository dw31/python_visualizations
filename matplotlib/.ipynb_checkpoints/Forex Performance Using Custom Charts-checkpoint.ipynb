```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Read in the data (from https://www.investing.com)

```python
forex = pd.read_csv('data-euro-usd.csv')
```

### Explore the data

```python
forex.info()
```

```python
forex.head()
```

```python
forex['Date'] = pd.to_datetime(forex['Date'])
forex.index = forex['Date']
del forex['Date']
forex.head()
```

### Plot Price

```python
plt.figure(figsize=(12, 6))
forex['Price'].plot()
plt.show()
```

### Plot % Change

```python
plt.figure(figsize=(12, 6))
forex['Change %'].plot()
plt.show()
```

### Plot 50 day moving average

```python
forex['Moving Average 50 Days'] = forex.Price.rolling(window=50).mean()
data = forex[['Price', 'Moving Average 50 Days']][-365:]
plots = data.plot(subplots=False, figsize=(12,5))
plt.show()
```

If we want to keep these separate

```python
plots = data.plot(subplots=True, figsize=(12,5))
plt.show()
```
